### 1.1 Python -- a language with high readability

Let's first imagine we want to make a number bomb game, a simple game that user guess a number with hint larger or smaller until it's correct. So let's fisrt write a **pseudocode** to implement a number bomb with a random answer from 1 to 100 to show our algorithm.

```py
secret_number = randint(1,100)

user_guess = input("Please enter a number: ")

while user_guess != secret_number:
    if secret_number > user_guess:
        print("Incorrect. The answer is larger")
    if secret_number < user_guess:
        print("Incorrect. The answer is smaller")
    # make a new guess
    user_guess = input("Please enter a number: ")

print("Correct")
```

And a Python implement looks like this

````python
secret_number = random.randint(1,100)

user_guess = int(input("Please enter a number: "))

while user_guess != secret_number:
    if secret_number > user_guess:
        print("Incorrect. The answer is larger")
    if secret_number < user_guess:
        print("Incorrect. The answer is smaller")
    # make a new guess
    user_guess = int(input("Please enter a number: "))

print("Correct")
````
