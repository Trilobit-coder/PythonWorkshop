#### Loops in Python

In Python, loops are mainly `for` loop or `while` loop. `for` loop can be vectorized like Matlab with `range()` function:

````python
# Basic for loop
for i in range(5):  # 0 to 4
    print(i)

# for loop with range parameters
for i in range(1, 6):  # 1 to 5
    print(i)

# for loop with step
for i in range(0, 10, 2):  # 0, 2, 4, 6, 8
    print(i)

for i in range(5, 0, -1):
    print(i)

# Iterating through a list （list is just array in Python)
fruits = ["apple", "banana", "cherry"]
for fruit in fruits:
    print(fruit)

# Iterating with index
for index, fruit in enumerate(fruits):
    print(f"{index}: {fruit}")
````

````python
# Basic while loop
count = 0
while count < 5:
    print(count)
    count += 1

# While loop with user input
password = ""
while password != "secret":
    password = input("Enter password: ")
print("Access granted!")

# Infinite loop with break
while True:
    user_input = input("Enter 'quit' to exit: ")
    if user_input == "quit":
        break
````

````python
# Loop Control Statements

# Break out of loop
for num in range(10):
    if num == 5:
        break
    print(num)  # Prints 0,1,2,3,4

# continue Statement
# Skip current iteration
for num in range(5):
    if num == 2:
        continue  # Skip number 2
    print(num)  # Prints 0,1,3,4

# pass Statement
# Placeholder statement
for num in range(5):
    if num % 2 == 0:
        pass  # Do nothing, but need something here
    else:
        print(f"Odd: {num}")

# else in loop
# else executes when loop completes normally (no break)
for num in range(5):
    print(num)
else:
    print("Loop completed successfully")

# else doesn't execute when break is used
for num in range(5):
    if num == 3:
        break
    print(num)
else:
    print("This won't print")
````

### 1.5 **Containers in Python**

In Python, containers are objects that hold other objects. The most common built-in containers are lists, tuples, dictionaries, and sets. Let's explore each one with examples.

#### 1.5.1 Lists - Mutable Ordered Collections
Lists are ordered, changeable, and allow duplicates. If you are coming from C++, list in python is not `std::list` in c++ but a dynamic array like `std::vector`. The memory in a list is linear.

````python
# Creating a list
fruits = ["apple", "banana", "cherry"]
numbers = [1, 2, 3, 4, 5]
mixed = ["text", 42, 3.14, True]

# Accessing elements
print(fruits[0])     # "apple" - first element
print(fruits[-1])    # "cherry" - last element

# Modifying lists
fruits.append("orange")      # Add to end
fruits.insert(1, "mango")    # Insert at position
fruits[2] = "blueberry"      # Change element
fruits.remove("banana")      # Remove specific item
popped = fruits.pop()        # Remove and return last item

# Slicing
print(numbers[1:3])   # [2, 3] - elements 1 to 2
print(numbers[:3])    # [1, 2, 3] - first 3 elements
print(numbers[2:])    # [3, 4, 5] - from index 2 to end

# List operations
combined = fruits + ["grape", "kiwi"]  # Concatenation
repeated = [1, 2] * 3                  # Repetition: [1, 2, 1, 2, 1, 2]
````

Tuples are ordered but cannot be changed after creation. You can view it as a `const` version of list.

````python
# Creating tuples
colors = ("red", "green", "blue")
single_item = ("only",)  # Note the comma for single-item tuple
coordinates = (10, 20)

# Accessing elements (same as lists)
print(colors[0])     # "red"
print(colors[-1])    # "blue"

# Tuples are immutable - this would cause an error:
# colors[0] = "yellow"  # TypeError!
````

#### 1.5.3 Dictionaries - Key-Value Pairs
Dictionaries store data as key:value pairs, unordered (Python 3.7+: insertion ordered). Since it is implemented with [hash table](https://en.wikipedia.org/wiki/Hash_table), it only take O(1) to look up value in dictionary through key (of course, if the hash distribution is good and not many collisions).

````python
# Creating dictionaries
person = {"name": "Alice", "age": 30, "city": "New York"}
scores = dict(math=95, science=88, history=92)  # Alternative syntax

# Accessing values
print(person["name"])        # "Alice"
print(person.get("age"))     # 30
print(person.get("email", "Not provided"))  # Default if key doesn't exist

# Modifying dictionaries
person["email"] = "alice@email.com"  # Add new key-value
person["age"] = 31                    # Update value
del person["city"]                    # Remove key
removed = person.pop("age")           # Remove and return value

# Dictionary methods
keys = person.keys()      # Get all keys
values = person.values()  # Get all values
items = person.items()    # Get all (key, value) pairs

# Looping through dictionaries
for key in person:
    print(key, person[key])

for key, value in person.items():
    print(f"{key}: {value}")
````

#### 1.5.4 Sets - Unordered Unique Collections
Sets store unique elements, unordered, and support mathematical set operations. Looking up element in a set also takes O(1) time because of hash table.

````python
# Creating sets
primes = {2, 3, 5, 7, 11}
even_numbers = set([2, 4, 6, 8, 10])  # Using set() constructor

# Sets automatically remove duplicates
colors = {"red", "green", "blue", "red", "green"}
print(colors)  # {'red', 'green', 'blue'}

# Set operations
A = {1, 2, 3, 4, 5}
B = {4, 5, 6, 7, 8}

print(A | B)   # Union: {1, 2, 3, 4, 5, 6, 7, 8}
print(A & B)   # Intersection: {4, 5}
print(A - B)   # Difference (in A but not B): {1, 2, 3}
print(A ^ B)   # Symmetric difference (in A or B but not both): {1, 2, 3, 6, 7, 8}

# Set methods
primes.add(13)          # Add element
primes.remove(2)        # Remove element (error if not exists)
primes.discard(100)     # Remove if exists (no error if not)
element = primes.pop()  # Remove and return arbitrary element
print(element, primes)
````

#### 1.5.5 Container Operations
All containers share some common operations:

#### 1.5.6 Comprehensions
Comprehensions are a short-cut to write containers with loops or conditions. Python provides comprehensions for lists, dictionaries, sets.

````python
# Basic comprehension
squares = [x**2 for x in range(10)]  # [0, 1, 4, 9, ..., 81]
print(squares)

# With condition
even_squares = [x**2 for x in range(10) if x % 2 == 0]
print(even_squares)

# Dictionary comprehension
square_dict = {x: x**2 for x in range(5)}  # {0: 0, 1: 1, 2: 4, 3: 9, 4: 16}
print(square_dict)

# Set comprehension
unique_lengths = {len(word) for word in ["hello", "world", "python"]}
print(unique_lengths)
````

#### 1.6.2 *args and **kwargs


`*`, in Python mean "unpacking". It allows to unpack the elments in a list or tuple. `*args` is arbitrary arguments. For examples, `*my_list` means unpack a list called my_list. With it, we can define a function with changeable amount of arguments.

````python
def make_pizza(*toppings):
    print("Making pizza with:")
    for topping in toppings:
        print(f"- {topping}")


make_pizza('mushrooms', 'peppers', 'cheese')
make_pizza('cheese')  # Works with any number
````


Similarly, `**` allows to unpack the elments in a dictionary. And `**kwargs` means arbitrary keyword arguments. It accepts any number of keyword arguments.

````python
def build_profile(first, last, **user_info):
    profile = {'first_name': first, 'last_name': last}
    for key, value in user_info.items():
        profile[key] = value
    return profile


user_profile = build_profile('John', 'Doe', age=30, city='NYC', job='Engineer')
for key, value in user_profile.items():
    print(f"{key}: {value}")
````

The `*args` and `**kwargs` is very critcal if you want to write a function with more flexibility, let's take a practice where we build a simple logger with `*args`

````python
def log_message(level, *messages):
    """Log multiple message parts"""
    full_message = "".join(str(msg) for msg in messages)
    print(f"[{str(level).upper()}] {full_message}")


connection_time = 100
log_message("error", "Failed to connect to database:", f"Connection timeout {connection_time}ms elapsed")
# Output: [ERROR] Failed to connect to database: Connection timeout
````

#### 1.6.3 Lifetime

The lifetime of the variables inside a function is different with those without a scope (in Python, `if` and `while` is not a scope). First, let's have a look of how to access and modify the global variable inside the function with `global` keyword.

````python
x = 10  # Global variable

def func1():
    x = 20  # This creates a new LOCAL variable
    print("Inside func1:", x)  # 20


func1()
print("Outside:", x)  # 10 (global unchanged)


def func2():
    global x  # Declare we're using the global x
    x = 20  # Modifies the global variable
    print("Inside func2:", x)  # 20


func2()
print("Outside:", x)  # 20 (global changed!)


def func3():
    # You can access the global variable without "global" keyword, but you can not modify it
    print("Inside func3:", x)


func3()
````

#### 1.6.4 [optional] Lambda

Lambda functions are anonymous functions. In Python, Lambda can be used with `lambda` keyword. The basic syntax is `lambda arguments: expression`. Lambda is very important in functional programming style.

````python
# Regular function
def add(x, y):
    return x + y


# Lambda equivalent
add_lambda = lambda x, y: x + y

print(add(3, 5))  # 8
print(add_lambda(3, 5))  # 8
````

````python
# Returns different values based on condition
grade_status = lambda score: "Pass" if score >= 60 else "Fail"
print(grade_status(75))  # returns "Pass"
print(grade_status(45))  # returns "Fail"

# Nested conditional (ternary)
classify = lambda x: (
    "positive" if x > 0 else
    "negative" if x < 0 else 
    "zero"
)

print(classify(5))  # returns "positive"
print(classify(-2))  # returns "negative"
print(classify(0))  # returns "zero"
````

#### 1.6.5 [optional] Generators

Generator function is the function that allow to save the state of the function and only return the a single value that is yielded. Once the value have been yielded, it will be paused right here until next called.

````python
def count_up_to(max):
    print("Generator Run")
    count = 1
    while count <= max:
        yield count  # Yields value, then pauses.
        print("Yield called")
        count += 1


counter = count_up_to(5)

print(next(counter))
print(next(counter), end="\n\n")

for num in counter:
    print(num)
````

#### 1.6.6 [optional] Decorators

A decorator is a callable (usually a function or a class) that:

- Takes a function as an argument

- Returns a (usually modified or wrapped) function

Decorators use this "function pointer" concept
When you write:

```python
@decorator
def my_func():
    pass
```
It's equivalent to:

```python
def my_func():
    pass
my_func = decorator(my_func)  # Passing function as argument
```

````python
def logger(func):
    def wrapper(*args, **kwargs):
        print(f"Calling '{func.__name__}{args}': ")
        return func(*args, **kwargs)
    return wrapper  # Returns a new "function pointer"


@logger  # this line is equiivalent to: add = logger(add), so now add is pointing to wrapper
def add(a, b):
    return a + b


# add is now pointing to wrapper, which internally calls the original add
print(add(2, 3))  # Prints: "Calling add" then 5
````

Decorators can add some new features (like logging, timing ...) while keeps the original functions unchanged. Let's make a timer that can count the running time of a function:

````python
def timer_decorator(func):
    def wrapper(*args, **kwargs):
        start = time.time()
        result = func(*args, **kwargs)
        end = time.time()
        print(f"{func.__name__} took {end-start:.4f} seconds")
        return result
    return wrapper

@timer_decorator
def slow_function(n):
    time.sleep(n)

slow_function(2)
````
